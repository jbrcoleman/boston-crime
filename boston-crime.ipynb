# Milestone 2
We will focus on continuing to clean the data and start to perform analysis on the data.

```python

crime_df.DAY_OF_WEEK = pd.Categorical(crime_df.DAY_OF_WEEK, 
              categories=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'],
              ordered=True)

crime_df.Lat.replace(-1, None, inplace=True)
crime_df.Long.replace(-1, None, inplace=True)

newcolnames = {'OFFENSE_CODE_GROUP':'Group',
         'OFFENSE_DESCRIPTION':'Description',
         'DISTRICT':'District',
         'REPORTING_AREA':'Area',
         'SHOOTING':'Shooting',
         'OCCURRED_ON_DATE':'Date',
         'YEAR':'Year',
         'MONTH':'Month',
         'DAY_OF_WEEK':'Day',
         'HOUR':'Hour',
         'STREET':'Street'}
crime_df.rename(index=str, columns=newcolnames, inplace=True)
```

# Find the most common types of crime

```python
import seaborn as sb
sb.catplot(y='Group',
           kind='count',
             height=10, 
            aspect=1.5,
            order=crime_df.Group.value_counts().index,
           data=crime_df)
```

# Explore Serious Crimes

Serious crime reported is categorized as UCR part One.

## Serious Crime Types

First we will perform the same analysis but only include serious crimes, UCR part one.

```python
serious_crime_df = crime_df.loc[crime_df['UCR_PART'] == 'Part One']

sb.catplot(y='Group',
           kind='count',
             height=8, 
            aspect=2,
            order=serious_crime_df.Group.value_counts().index,
           data=serious_crime_df)
```

It looks like larceny is the most common serious crime type reported in Boston. Larceny and larceny from motor vehicle account for the majority of major crimes reported. 

## When does major crime happen?

For analyzing when major crimes happen we will only analyze data from 2016 and 2017 since they are the only full years of data we have. 

```python
serious_crime_1617_df = serious_crime_df.loc[serious_crime_df['Year'].isin([2016,2017])]
```

```python
# Crimes by hour of the day
sb.catplot(x='Hour',
           kind='count',
            height=8, 
            aspect=2,
            color='gray',
           data=serious_crime_1617_df)
```

From the data we can see that there is a considerable dropoff in the amount of serious crime that happens after midnight until the morning.

```python
sb.catplot(x='Day',
           kind='count',
            height=6, 
            aspect=2,
           data=serious_crime_1617_df)
```

```python
sb.catplot(x='Month',
           kind='count',
            height=8, 
            aspect=2,
            color='orange',
           data=serious_crime_1617_df)
```

# Where does major crime happen?



Now we will take look at where major crime happens in Boston durinig 2016 and 2017. Below is a map of the policing district in boston and we will use this as a reference to break down where major crime happens in the city. 

```python
# Serious crimes by District
import matplotlib.pyplot as plt
sb.catplot(x='District',
           kind='count',
            height=8, 
            aspect=3,
           data=serious_crime_1617_df)
plt.xticks(size=30)
plt.yticks(size=30)
plt.xlabel('')
plt.ylabel('Count', fontsize=40)
```

It looks like the majority of serious crimes occur in the heart of the city with the districts having the most serious crimes reported being districs D4, B2, and A1.The amount of serious crimes tapers off in the outskirts of the city.

```python
# Scatterplot broken down by district showing where using latitude and longitude to show where serious crime happens.
plt = sb.scatterplot(x='Long',
               y='Lat',
                alpha=.02,
                hue='District',
               data=serious_crime_1617_df)
plt.legend(loc='center left', bbox_to_anchor=(1.25, 0.5), ncol=1)
```

Here shows a scatter plot seperated by districts to help better illustrate the distribution of crime in Boston. This graph also reinforces the greater density of serious crime in the heart of the city (A1,B2, and D4), and the tapering off of serious crime toward the suburbs of Boston.

## Larceny Heatmap

#### We will look at a heatmap of where the most common serious, Larceny, occured. 

```python
import folium
from folium.plugins import HeatMap
larceny_df = serious_crime_1617_df.loc[serious_crime_1617_df['Group'] == 'Larceny']

# Create basic Folium crime map
base_map = folium.Map(location=[42.23, -71.17],
                      tiles = 'openstreetmap',
                      zoom_start = 11)

# Clean data for heatmap 
larceny_heatmap = larceny_df[['Lat','Long']]
larceny_heatmap = larceny_df.dropna(axis=0, subset=['Lat','Long'])
larceny_heatmap = [[row['Lat'],row['Long']] for index, row in larceny_heatmap.iterrows()]
HeatMap(larceny_heatmap, radius=10).add_to(base_map)

base_map
```

### Comparing serious Crime incidents from 2016 to 2017

```python
sb.catplot("Year",  col="Group",
                data=serious_crime_1617_df, saturation=.5,
                kind="count", ci=None, aspect=.8, col_wrap=3)
```

#### Crime rates are low past midnight till early morning. The rate gradually increases towards the evening and appears to peak at 6 pm. Variation across days of the week doesn't seem much, though Friday has the highest crime rate. During the year, summer months ( June through August ) have a high rate compared to rest of the year.  January is an outlier that has a high crime rate. 

#### Serious Crimes are the highest in A1 and D4 districts as indicated by Lat/Long graph. These are the downtown and south Boston districts. 

#### There is a slight decrease in serious crime rate from 2016 to 2017. Number of Burglaries have decreased and there were fewer homicides.  Larceny from motor vehicle ( the largest nbr of incidents) doesn't changed much.
